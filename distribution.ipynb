```python
import csv
import numpy as np
import matplotlib.pyplot as plt
file_path = 'integers.csv'
values = []
with open(file_path, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader, None)
    for row in csvreader:
        try:
            value = float(row[1])
            if np.isfinite(value):
                values.append(value)
            else:
                print(f"Non-finite number found and skipped: {row[1]}")
        except ValueError as e:
            print(f"Error converting '{row[1]}' to float: {e}")

if not values:
    raise ValueError("No valid numeric data found in the file.")

data = np.array(values)
average = np.mean(data)
minimum = np.min(data)
maximum = np.max(data)
std_dev = np.std(data)
print(f"Average: {average}")
print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
print(f"Standard Deviation: {std_dev}")
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, color='blue', alpha=0.7, density=True)
plt.title('Probability Distribution of Values')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.axvline(average, color='red', linestyle='dashed', linewidth=2)
plt.text(average, plt.ylim()[1]*0.9, f'Mean: {average:.2f}', color = 'red')
plt.show()
```
